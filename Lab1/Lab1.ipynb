```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from PIL import Image
```

### Task #2: Loading and Visualizing Images:


#### a. Loading/reading image using OpenCv

```python
img = cv2.imread("cat.jpg")
# In this case, the image dose not appear with its real colors because OpenCV reads images using BGR color space, whereas Matplotlib displays images using RGB color space.(This issues could be fixed using cvtColor function to convert the color space)
plt.imshow(img)
plt.show
```

```python
# fix the color by convert the colors space from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show
```

#### b. Loading/reading image using PIL

```python
img2 = Image.open("dog.jpg")
plt.imshow(img2, cmap = cm.Greys)
plt.show()
```

### Task #3: Image Storing

#### a. Saving image to disc using OpenCV

```python
cv2.imwrite("save_cat_image.jpg",img)
```

#### b.	Saving image to disc using PIL

```python
img2.save("save_dog_image.jpg")
```
